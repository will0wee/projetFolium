```python
dataFrame = pandas.DataFrame(data = le_fooding_results)

for i in range(len(dataFrame.index)):
    dfAddress = pandas.DataFrame(data = le_fooding_results[i].get('address'))
    dfAddress['coordonne'] = str([dfAddress.iat[0, 5], dfAddress.iat[1, 5]])
    dfAddress.drop(columns='position', axis=1)
    dataFrame.iat[i, 3] = dfAddress.loc[0]
    
    tmp = le_fooding_results[i].get('sorted_tags')
    
    for elem in tmp.keys():
        if(isinstance(tmp.get(elem), list)):
            tmp[elem] = ', '.join(tmp.get(elem))
        
    dftags = pandas.DataFrame(data = tmp, index = [0])
    dataFrame.iat[i, 12] = dftags
    
    for key in dataFrame.keys():
        if(isinstance(dataFrame.loc[i, key], list)):
            verif = True
            for elem in dataFrame.loc[i, key]:
                if(isinstance(elem, dict)):
                    verif = False
            if(verif):
                dataFrame.loc[i, key] = ', '.join(dataFrame.loc[i, key])
            
#dataFrame.loc[0,'cravings']
le_fooding_results[0].get('sorted_tags')
```
